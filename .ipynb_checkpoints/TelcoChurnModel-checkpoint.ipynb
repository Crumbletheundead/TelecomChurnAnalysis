## Data Wrangling

```python
#Condense every data wrangling into one section
rawData = pd.read_csv('TelcoCustomerChurn.zip')
rawData.TotalCharges = rawData.TotalCharges.apply(lambda x: 0 if ' ' in x else float(x))
```

```python
booleanCols = [
    'SeniorCitizen','Partner','Dependents','PhoneService','MultipleLines',
    'OnlineSecurity','OnlineBackup','DeviceProtection',
    'TechSupport','StreamingTV','StreamingMovies','PaperlessBilling'
]

catCols = [
    'InternetService','PaymentMethod'
]

numericCols = [
    'tenure','MonthlyCharges','TotalCharges'
]
ordinalCols=[
    'Contract'
]
targetCol = ['Churn']
```

```python
def strBoolConverter(val):
    """Maps columns that encode booleans with
    string dtypes to actual booleans
    'No' -> False
    'Yes' -> True
    'No XXX' -> np.nan
    """
    if isinstance(val,int):
        return val
    elif 'No ' in val:
        return np.nan
    elif 'No' in val:
        return False
    else:
        return True

def isFemale(val):
    """converts gender column from string to bool"""
    if 'Female' in val:
        return True
    else:
        return False

def ordContract(val):
    """Maps Contract length strings into ordinal
    variables, where 
    'month-to-month' -> 0
    'one' -> 1
    'two' -> 2
    """
    if 'Month' in val:
        return 0
    elif 'One' in val:
        return 1
    elif 'Two' in val:
        return 2
    else:
        return np.nan

    
colBConverter = np.frompyfunc(strBoolConverter,1,1)
colIsFemale = np.frompyfunc(isFemale,1,1)
colOrdContract= np.frompyfunc(ordContract,1,1)
```

```python
colTransform = make_column_transformer(
    ('passthrough',['customerID','SeniorCitizen']),
    (FunctionTransformer(colIsFemale),['gender']),
    (FunctionTransformer(colOrdContract),['Contract']),
    (FunctionTransformer(colBConverter), booleanCols),
    (MinMaxScaler(), numericCols),
    (OneHotEncoder(), catCols),
    (FunctionTransformer(colBConverter),['Churn'])
)
```

```python
processedData = colTransform.fit_transform(rawData)
```

```python
train,test = train_test_split(processedData,test_size=0.2,random_state=42)
```

StratifiedKfold for model selection

```python
skf = StratifiedKFold(n_splits=5,shuffle=True,random_state=1337)
modelTestFolds = []
for foldTrain,foldValid in skf.split(train[:,1:-1],train[:,-1].astype(bool)):
    modelTestFolds.append((train[foldTrain],train[foldValid]))
```

```python
trainX = train[:,1:-1].astype(float)
trainY = train[:,-1].astype(bool)
testX = test[:,1:-1].astype(float)
testY = test[:,-1].astype(bool)
```

# Model Training and Hyperparameter Tuning

## XGBoost

```python
xgbClf = xgb.XGBClassifier(random_state=1337,use_label_encoder=False)
```

### Grid search high range of group1 params
#### Group 1 Params
* max_depth
* min_child_weight
* subsample
* colsample_bytree

```python
xgbParams = {
    'learning_rate':[0.1],
    'max_depth': [4,8,10],
    'min_child_weight':[1,5,10],
    'subsample':[1,0.9,0.8],
    'colsample_bytree':[1,0.8],
    'n_estimators':[100],
    'objective': ['binary:logistic'],
    'scale_pos_weight':[(~trainY.astype(bool)).sum()/trainY.sum()],
    'eval_metric':['auc']
}
```

```python
grid_search = GridSearchCV(xgbClf, param_grid=xgbParams,scoring='roc_auc',verbose=1)
xgbGridResult = grid_search.fit(trainX,trainY)
```

```python
group1Params = ['param_max_depth','param_min_child_weight','param_subsample','param_colsample_bytree']
```

```python
cvResHigh = pd.DataFrame(xgbGridResult.cv_results_)
cvResHigh.sort_values('rank_test_score',ascending=True)[
    group1Params +
    ['mean_test_score','rank_test_score']
].head(5)
```

### Grid search LR + Number of Boost Rounds

```python
xgbParams = {
    'learning_rate':[0.1,0.01],
    'max_depth': [4],
    'min_child_weight':[10],
    'subsample':[1],
    'colsample_bytree':[0.8],
    'n_estimators':[50,100,150,200],
    'objective': ['binary:logistic'],
    'scale_pos_weight':[(~trainY.astype(bool)).sum()/trainY.sum()],
    'eval_metric':['auc']
}
```

```python
grid_search = GridSearchCV(xgbClf, param_grid=xgbParams,scoring='roc_auc',verbose=1)
xgbGridResult = grid_search.fit(trainX,trainY)
```

```python
cvResLr = pd.DataFrame(xgbGridResult.cv_results_)
cvResLr.sort_values('rank_test_score',ascending=True)[[
    'param_learning_rate','param_n_estimators',
    'mean_test_score','rank_test_score'    
]].head(5)
```

```python
sns.scatterplot(data=cvResLr,x='param_n_estimators',y='mean_test_score',hue='param_learning_rate')
```

### Check best candidate against testing set

```python
fpr,tpr,thresh= roc_curve(testY,xgbGridResult.best_estimator_.predict_proba(testX)[:,1])

plt.figure(figsize=[4,4])
ax = sns.lineplot(x=fpr,y=tpr)
sns.lineplot(x=[0,1],y=[0,1],ax=ax)
ax.set_title('AUC = {}'.format(roc_auc_score(testY,xgbGridResult.best_estimator_.predict_proba(testX)[:,1])))
```
